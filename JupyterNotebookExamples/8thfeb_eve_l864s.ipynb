```python
#train test split
from sklearn.model_selection import train_test_split
```

```python
trainx, testx, trainy, testy = train_test_split(X,y, test_size=0.3, random_state=1000)
```

```python
#import ml algo - linear regression
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linregmodel = LinearRegression()
```

```python
#train model using fit function
```

```python
linregmodel.fit(trainx, trainy)
```

```python
linregmodel.intercept_
```

```python
linregmodel.coef_
```

```python
linregpred_op = linregmodel.predict(testx)
```

```python
comparedf=pd.DataFrame({'Test Y':testy, 'predictedop':linregpred_op})
```

```python
comparedf['error']=comparedf['Test Y']-comparedf['predictedop']
```

```python
comparedf
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(testy, linregpred_op)
```

```python
sqrt(mean_squared_error(testy, linregpred_op))
```
